```python
with open("compound_score_wise_after_cal_20000.json") as f:
    cs= json.loads(f.read())
```

```python
def get_score(senti_frame,auth):
    f=senti_frame.loc[senti_frame["Author"]==auth]
    summ=0
    for i in f["compound"]:
        summ+=i
    return summ/len(f["compound"])
                         
```

```python
#calculating the edges
edges={}
edge_confidence={}
for i in cs.keys():
    try:
        edges[i]=mean(cs[i])
        edge_confidence[i]=len(cs[i])
    except:
        continue
    
```

```python
#Given an author name returns the apps in all which they reviewed
def get_apps(Author):
    with open("info_frequency.csv","r", encoding="utf8") as f:
        reader = csv.reader(f)
        num=0
        for row in reader:
            if num==0:
                num+=1
                continue
            if row[0]==Author:
                num+=1
                apps=app_name_split(row[2])
                break
    return apps
        
```

```python
def get_s_score(Author,app):
    try:
        senti_frame=pd.read_csv("datasets_sentiment\\sentiment\\"+"Sentiment_Score_for_Reviews_of_"+app+".csv")
        return get_score(senti_frame,Author)
    except:
        print("prob_with_ge_s_score")
        return None
```

```python
def author_scores(Author):
    Author_scores={}
    apps=get_apps(Author)
    for app in apps:
        t=get_s_score(Author,app)
        if not(t==None):
            Author_scores[app]=get_s_score(Author,app)
        else:
            continue
    return Author_scores

def sentiment_predictor(app1,s_score,app2):
    try:
        if edge_confidence[app1+":"+app2]>=5:#checking for confidence
            diff=edges[app1+":"+app2]
            return s_score+diff
        else:
            print("No enough confidence on------ "+ app1+":"+app2 )
            return None
    except:
        print("pair data not available")
        return None
def get_pred_scores(aus):
    split_index=int(len(aus.keys())/2)
    train_apps=list(aus.keys())[:split_index]
    test_apps=list(aus.keys())[split_index:]
    pred_scores={}
    for te_app in test_apps:
        pred_scores[te_app]=[]
        for tr_app in train_apps:
            st=sentiment_predictor(tr_app,aus[tr_app],te_app)
            if not(st==None):
                pred_scores[te_app].append(st)   
            else:
                continue
                
    pred_mean={}
    for i in pred_scores.keys():
        try:
             pred_mean[i]=mean(pred_scores[i])
        except:
            continue
    return pred_mean
def mean_sq_err(pred,aus):
    ytrue=[]
    for i in pred.keys():
        ytrue.append(aus[i])
        
    return mean_squared_error(ytrue,list(pred.values()))
```

```python
#testing the prediction system for author Matt Arnold
aus=author_scores('Matt Arnold')
predicted=get_pred_scores(aus)
if len(predicted)>0:
    print(mean_sq_err(predicted,aus))
else:
    "No confidence to predict any scores"
```
